# **Fraud Check Dataset**

```python
# Importing the libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

There are no null values in the dataset.

```python
# Scaling Data

fraud_check['Taxable.Income']=np.where(fraud_check['Taxable.Income']<=30000,1,0)
```

```python
fraud_check
```

```python
fraud_check['Taxable.Income'].value_counts()
```

```python
fraud_check=pd.get_dummies(fraud_check,columns=['Undergrad','Marital.Status','Urban'],drop_first=True)
```

```python
fraud_check
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
fraud2=norm_func(fraud_check)
```

```python
fraud2
```

```python
x=fraud2.iloc[:,1:]
y=fraud2.iloc[:,0]
```

```python
x
```

```python
y
```

```python
# Splitting data into training and testing Data set.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=90)
```

# **Building Decision Tree Classifier using Entropy Criteria**

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
model.fit(x_train,y_train)
```

```python
tree.plot_tree(model);
```

```python
# Plotting Decision Tree

fig=plt.figure(figsize=(25,20))
fig=tree.plot_tree(model,
feature_names=['Undergrad','Martial.Status','City.Population','Work.Experience'],class_names=["Risky","Good"],filled=True)
plt.title('Decision Tree using Entropy',fontsize=22);
```

# **Prediction**

```python
predict=model.predict(x_test)
```

```python
pd.crosstab(y_test,predict)
```

```python
y_test.value_counts()
```

```python
# Accuracy 

from sklearn.metrics import accuracy_score

accuracy_score(y_test,predict)
```

```python
model2 = DecisionTreeClassifier(criterion = 'entropy', max_depth=4)
model2.fit(x_train,y_train)
```

```python
predict=model2.predict(x_test)
```

```python
accuracy_score(y_test,predict)
```

# **Building a model using Gini Index.**

```python
from sklearn.tree import DecisionTreeClassifier
model_gini1 = DecisionTreeClassifier(criterion='gini', max_depth=2)
model_gini1.fit(x_train,y_train)
```

```python
#Prediction and computing the accuracy
pred=model.predict(x_test)
np.mean(predict==y_test)
```
